```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', font_scale=1.1)
```

```python
# Load the dataset
df = pd.read_csv('population_data.csv')
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head()
```

```python
# Clean the dataset
df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]
df.drop_duplicates(inplace=True)
df = df.dropna(subset=['county', 'city', 'make', 'model', 'model_year'])
```

```python
# Display EV adoption trends by year
ev_by_year = df.groupby('model_year').size().reset_index(name='vehicle_count')
plt.figure(figsize=(10,6))
sns.lineplot(x='model_year', y='vehicle_count', data=ev_by_year, marker='o')
plt.title('EV Registrations in Washington State by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Number of EVs Registered')
plt.show()
```

```python
# Top countries for EV Adoption (Latest Year)
latest_year = df['model_year'].max()
top_counties = df[df['model_year']==latest_year].groupby('county').size().reset_index(name='ev_count')
top_counties = top_counties.sort_values('ev_count', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='ev_count', y='county', data=top_counties)
plt.title(f'Top 10 Counties by EV Registrations ({latest_year})')
plt.xlabel('Number of EVs Registered')
plt.ylabel('County')
plt.show()
```

```python
# EV Type Breakdown 
if 'electric_vehicle_type' in df.columns:
    type_counts = df['electric_vehicle_type'].value_counts()
    plt.figure(figsize=(8,5))
    type_counts.plot(kind='bar')
    plt.title('EV Types Registered')
    plt.ylabel('Count')
    plt.xlabel('EV Type')
    plt.show()
```

```python
# Display top makes/models (Latest Year)
top_models = df[df['model_year']==latest_year].groupby(['make','model']).size().reset_index(name='count')
top_models = top_models.sort_values('count', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='count', y=top_models['make'] + ' ' + top_models['model'], data=top_models)
plt.title(f'Top 10 EV Makes/Models Registered in {latest_year}')
plt.xlabel('Number Registered')
plt.ylabel('Make & Model')
plt.show()
```

```python
# Electric Range Distribution
if 'electric_range' in df.columns:
    plt.figure(figsize=(10,6))
    sns.histplot(df['electric_range'].dropna(), bins=30)
    plt.title('Distribution of Electric Vehicle Range (miles)')
    plt.xlabel('Electric Range (miles)')
    plt.ylabel('Number of Vehicles')
    plt.show()
```
